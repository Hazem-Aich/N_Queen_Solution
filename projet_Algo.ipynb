```python
import copy
import random



def saisie():
 
    while True:
        try:
            n = int(input('Donner le nombre de Reines ? n = '))
            if n <= 3:
                print("donner une valeur superieur ou égal  4")
                continue
            return n
        except ValueError:
              print("valeur n'est pas valide !! ")
n = saisie()

def Remplir(n):
    
    T = [" "]*n
    for i in range(n):
        T[i] = [" "]*n
    return T

def Afficher_solution(solutions, n):
 
    x = random.randint(0,len(solutions)-1)

    for row in solutions[x]:
        
        print("| " + " | " .join(row) +" | ")




def Est_Libre(T, row, col, n):

    for j in range(col):
        if T[row][j] == "R":
            return False
    
    i, j = row, col
    while i >= 0 and j >= 0:
        if T[i][j] == "R":
            return False
        i=i-1
        j=j-1
    
    x, y = row,col
    while x < n and y >= 0:
        if T[x][y] == "R":
            return False
        x=x+1
        y=y-1
    
    return True

def ajouter_solution(T):

    global solutions
    saved_board = copy.deepcopy(T)
    solutions.append(saved_board)


    
T = Remplir(n)

solutions = []

def placer_reine(T, col, n):
    
    if col >= n:
        return
    
    for i in range(n):
        if Est_Libre(T, i, col, n):
            T[i][col] = "R"
            if col == n-1:
                ajouter_solution(T)
                T[i][col] = " "
                
            placer_reine(T, col+1, n) 
            T[i][col] = " "
        
placer_reine(T, 0, n)
print()
Afficher_solution(solutions, n)
print()
p=len(solutions)
print("Nombre de solutions totale est ègal  "  + (str)(p)+ "")
```
